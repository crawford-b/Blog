```python
states = alt.topo_feature(data.us_10m.url, 'states')

# US states background
background = alt.Chart(states).mark_geoshape(
    fill='#edf5f9',
    stroke='white'
).properties(
    title='US Mass Shootings',
    width=650,
    height=400
).project('albersUsa')

background
# Adding a selection
selection = alt.selection_single(on="mouseover", fields=['year'])

# US Shootings points
shootings = alt.Chart().mark_circle(size=10).encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    color=alt.Color('venue:N', scale=alt.Scale(domain=venues), legend=alt.Legend(title = "Venue")),
    size=alt.Size('fatalities:Q', scale=alt.Scale(domain=(5, 50))),
    tooltip='case:N'
).transform_filter(
    selection
)

#shootings.encode(alt.Color('Miles_per_Gallon', scale=alt.Scale(domain=venues)))

top = background + shootings 

# #------------ BOTTOM ----------------#

bottom = alt.Chart().mark_bar().encode(
    x='year:N',
    y=alt.Y('sum(fatalities):Q', title="Total fatalities"),
    color=alt.condition(selection, alt.value('steelblue'), alt.value('lightgray'))
).properties(
    width=650, height=100,
    selection=selection
)

alt.vconcat(
    top, bottom,
    data=shooting_csv
)
```

```python
brush = alt.selection(type='interval')

points = alt.Chart().mark_point().encode(
    x = alt.X('date:T', timeUnit='year'),
    y = 'fatalities:Q',
    color = alt.condition(brush, 'venue:N', alt.value('lightgray'))
).properties(
    selection=brush
)

bars = alt.Chart().mark_bar().encode(
    y='venue:N',
    color=alt.Color('venue:N', legend=None),
    x=alt.X('sum(fatalities):Q', title="Total fatalities")
).transform_filter(
    brush.ref()
)

alt.vconcat(points, bars, data=shooting_csv)
```

```python
shooting['year'] = shooting['date'].map(lambda x: x.year)
subdf = shooting.groupby('year')['fatalities'].agg({'total_fatalities':sum, 'num_shooting':'count'}).reset_index()
subdf['idx'] = subdf['year'] - subdf.year.min()

max_idx = subdf.idx.max()
all_idx = subdf.idx.unique()
min_year = subdf.year.min()

for y in range(max_idx):
    if y not in all_idx: 
        subdf = subdf.append({'idx':y, 'year':y+min_year, 'total_fatalities':0, 'num_shooting':0}, ignore_index=True)

subdf = subdf.sort_values('idx')
```

```python
n_years = len(subdf)
years = subdf.idx.values
fatality_data = subdf.total_fatalities.values
shootings_data = subdf.num_shooting.values
```

# Total number of fatalities

```python
import pymc3 as pm
from pymc3.math import switch

with pm.Model() as us_mass_fatal:
    
    early_mean = pm.Exponential('early_mean', 1)
    late_mean = pm.Exponential('late_mean', 1)
    
    switchpoint = pm.DiscreteUniform('switchpoint', lower=0, upper=n_years)
    
    rate = switch(switchpoint >= np.arange(n_years), early_mean, late_mean)
    
    shootings = pm.Poisson('shootings', mu=rate, observed=fatality_data)
    
```

```python
with us_mass_fatal:
    trace_fatal = pm.sample(10000, tune=1000)
```

```python
_ = pm.traceplot(trace_fatal)
```

```python
plt.hist(trace_fatal['switchpoint'])
```

```python
pm.forestplot(trace_fatal)
```

```python
cmap = matplotlib.cm.get_cmap('Purples')
normed = matplotlib.colors.Normalize(vmin=0, vmax=subdf.total_fatalities.max())
colors = [cmap(normed(x)) for x in subdf.total_fatalities]

f, ax = plt.subplots(figsize=(10,4))
sns.barplot(subdf.year, subdf.total_fatalities, ax=ax, palette=colors)
ax.set_xticklabels(ax.get_xticklabels(), rotation=60)

sp = trace_fatal['switchpoint'].mean()

early_rate = trace_fatal['early_mean'].mean()
late_rate = trace_fatal['late_mean'].mean()


yl, yh = ax.get_ylim()
xl, xh = ax.get_xlim()

ax.axhline(y = early_rate, xmin=0, xmax = sp/xh , ls="--", lw=2,color='tomato')
ax.axvline(x = sp, ymin = early_rate/yh, ymax = late_rate/yh, ls="--", lw=2,color='tomato')
ax.axhline(y = late_rate, xmin=sp/xh, xmax = 1, ls="--", lw=2, color='tomato')

rand_idx = np.random.choice(np.arange(10000), 500)

for i in rand_idx:
    
    rate1 = trace_fatal['early_mean'][i]
    rate2 = trace_fatal['late_mean'][i]
    
    switch = trace_fatal['switchpoint'][i]
    
    if switch > 27: color = 'dodgerblue'
    else: color = 'tomato'
        
    ax.axhline(y = rate1, xmin=0, xmax = switch/xh , alpha= 0.05, lw=0.5, color=color)
    ax.axvline(x = switch, ymin = rate1/yh, ymax = rate2/yh, alpha= 0.05, lw=0.5, color=color)
    ax.axhline(y = rate2, xmin=switch/xh, xmax = 1, alpha= 0.05, lw=0.5, color=color)

sns.despine(left=True)
ax.set_xlabel("")
ax.set_ylabel("Total Fatalities")

plt.savefig('2018-05-22_fatalities.png', dpi=250)
```

```python
import pymc3 as pm
from pymc3.math import switch

with pm.Model() as us_mass_sh:
    
    early_mean = pm.Exponential('early_mean', 1)
    late_mean = pm.Exponential('late_mean', 1)
    
    switchpoint = pm.DiscreteUniform('switchpoint', lower=0, upper=n_years)
    
    rate = switch(switchpoint >= np.arange(n_years), early_mean, late_mean)
    shootings = pm.Poisson('shootings', mu=rate, observed=shootings_data)
    
    
    
```

```python
with us_mass_sh:
    trace = pm.sample(10000, tune=1000)
```

```python
pm.traceplot(trace)
```

```python
pm.forestplot(trace)
```

```python
low_switch = np.where(trace['switchpoint'] < 27)
high_switch = np.where(trace['switchpoint'] > 27)
```

```python
switch_low = trace['switchpoint'][low_switch].mean()
switch_high = trace['switchpoint'][high_switch].mean()
```

```python
plt.hist(trace['early_mean'][low_switch], alpha=0.3, bins=20)
_ = plt.hist(trace['early_mean'][high_switch], alpha=0.3, bins=20)
rate1_low = trace['early_mean'][low_switch].mean()
rate1_high = trace['early_mean'][high_switch].mean()

plt.hist(trace['late_mean'][low_switch], alpha=0.3, bins=20)
_ = plt.hist(trace['late_mean'][high_switch], alpha=0.3, bins=20)
rate2_low = trace['late_mean'][low_switch].mean()
rate2_high = trace['late_mean'][high_switch].mean()
```

```python
sns.set_style('whitegrid')
f, ax = plt.subplots(1, 2, figsize=(10, 5), sharex=True, sharey=True)
ax[0].hist(trace['early_mean'][low_switch], density =True, color='dodgerblue', 
           alpha=0.3, bins=20, label=r'early rate: $\lambda_1$')
ax[0].hist(trace['late_mean'][low_switch], density =True, color='tomato', 
           alpha=0.3, bins=20, label=r'late rate: $\lambda_2$')
ax[0].axvline(trace['early_mean'][low_switch].mean(), ls='--', color='dodgerblue')
ax[0].axvline(trace['late_mean'][low_switch].mean(), ls='--', color='tomato')
ax[0].set_title("Lower mode", fontsize=14)
ax[0].legend()

ax[1].hist(trace['early_mean'][high_switch], density =True, color='dodgerblue', 
           alpha=0.3, bins=20, label=r'early rate: $\lambda_1$')
ax[1].hist(trace['late_mean'][high_switch], density =True, color='tomato', 
           alpha=0.3, bins=20, label=r'late rate: $\lambda_2$')
ax[1].axvline(trace['early_mean'][high_switch].mean(), ls='--', color='dodgerblue')
_ = ax[1].axvline(trace['late_mean'][high_switch].mean(), ls='--', color='tomato')
ax[1].set_title("Higher mode", fontsize=14)
ax[1].legend()
sns.despine(left=True)

plt.savefig('2018-05-22_mass_shootings.png', dpi=250)
```
