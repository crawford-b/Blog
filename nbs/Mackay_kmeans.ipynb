# Simple k-means

```python
def assigment_step(centroids, xs):
    
    assignments = []
    
    for x in xs:
        best_centroid = 0
        best_distance = np.inf
        for c_id, c in enumerate(centroids):
            x_to_c = sp.spatial.distance.euclidean(x, c)
            if x_to_c < best_distance:
                best_centroid = c_id
                best_distance = x_to_c
        
        assignments.append(best_centroid)
    
    return np.array(assignments)
```

```python
def update_step(xs, assignments, n):
    
    all_centroid = []
    
    for i in range(n):
        i_idx = np.argwhere(assignments == i).squeeze()
        assigned_xs = xs[i_idx].reshape(-1, 2)
        centroid = np.sum(assigned_xs, axis = 0) / assigned_xs.reshape(-1, 2).shape[0]
        all_centroid.append(centroid)
    return np.row_stack(all_centroid)
```

```python
def make_kmeans_plot(xs, assignments, centroids, n_clusters, sigmas = [], ambigious = [], ax=None):

    cmap = mpl.cm.get_cmap('Set1')
    colors = np.array(cmap.colors)

    if ax == None:
        f, ax = plt.subplots(figsize=(7,7))
        ax.set_aspect('equal', adjustable='box')
    if len(sigmas) == 0:
        sigmas = np.full_like(centroids, 1)
        print("no sigma specified")
    if len(ambigious) > 0:
        ax.scatter(xs[ambigious, 0], xs[ambigious, 1], facecolors='none', edgecolors='dodgerblue')
        xs = np.delete(xs, ambigious, axis=0)
        assignments = np.delete(assignments, ambigious) 
        
    ax.scatter(xs[:,0], xs[:,1], marker="x", c=colors[assignments])
    ax.scatter(centroids[:,0], centroids[:,1], s=20, color=colors[np.arange(n_clusters)])

    all_circ = []
    for i, c in enumerate(centroids):
        sig = sigmas[i]
        if len(sig.shape) == 2:
            
            lambda_, v = np.linalg.eig(sig)
            lambda_ = np.sqrt(lambda_)

            x_sig = lambda_[0] * 2
            y_sig = lambda_[1] * 2            
            
            angle = np.degrees(np.arctan2(v[1,0], v[0,0]))
            
            print("Covariance specified")
        else:
            x_sig = 2*sig[0]
            y_sig = 2*sig[1]
            angle = 0
            
        #circ = mpl.patches.Circle(c, sig, color=colors[i], alpha=0.2, label=i)
        circ = mpl.patches.Ellipse(c, x_sig, y_sig, angle=angle, color=colors[i], alpha=0.2, label=i)
        all_circ.append(circ)
        ax.add_artist(circ)
    
    #ax.legend(handles=all_circ)
    
    return ax
```

```python
def run_simulation(xs, n_clusters):
    
    # We don't want to intialize two clusters to the same point
    unique_xs = np.unique(xs, axis=0)

    # initialize clusters
    centroids_idx = np.random.choice(np.arange(unique_xs.shape[0]), size=n_clusters, replace=False)
    new_centroids = unique_xs[centroids_idx]
    prev_centroids = np.zeros_like(new_centroids)

    assignments = assigment_step(new_centroids, xs)
    iteration = 0
    
    while not np.isclose(new_centroids, prev_centroids, atol=1e-10, rtol=1e-8).all():
        
        prev_centroids = new_centroids
        new_centroids = update_step(xs, assignments, n_clusters)
        assignments = assigment_step(new_centroids, xs)
        
        iteration += 1
    
    assignments2 = np.invert(assignments, dtype=np.int8) + 2
    assignments = np.column_stack([assignments2, assignments])

    return assignments, new_centroids, []
```

# Soft k-means clustering

```python
def softmax(arr, axis=1):
    
    return np.exp(arr) / np.exp(arr).sum(axis=axis).reshape(-1, 1)
```

```python
def assignment_step_soft(centroids, xs, beta):
    
    assignments = []
    
    distance_to_centroids = np.zeros((xs.shape[0], centroids.shape[0]))
    
    for x_id, x in enumerate(xs):
        for c_id, c in enumerate(centroids):
            distance_to_centroids[x_id, c_id] = -beta * sp.spatial.distance.euclidean(x, c)
            
    return softmax(distance_to_centroids, axis=1)
```

```python
def update_step_soft(xs, assignments, n):
    
    all_centroid = []
    
    for i in range(n):
        centroid_resps = assignments[:, i]
        centroid = centroid_resps @ xs / centroid_resps.sum()
        all_centroid.append(centroid)
    
    return np.row_stack(all_centroid)
```

```python
def run_simulation_soft(xs, n_clusters, beta = 2):
    
    # We don't want to intialize two clusters to the same point
    unique_xs = np.unique(xs, axis=0)

    # initialize clusters
    centroids_idx = np.random.choice(np.arange(unique_xs.shape[0]), size=n_clusters, replace=False)
    new_centroids = unique_xs[centroids_idx]
    prev_centroids = np.zeros_like(new_centroids)

    assignments = assignment_step_soft(new_centroids, xs, beta)
    iteration = 0
    
    while not np.isclose(new_centroids, prev_centroids, atol=1e-12, rtol=1e-10).all():
        
        prev_centroids = new_centroids
        new_centroids = update_step_soft(xs, assignments, n_clusters)
        assignments = assignment_step_soft(new_centroids, xs, beta)
        
        iteration += 1
        
    return assignments, new_centroids, []
```

# Soft K-means clustering v2

```python
def gaussian_dist(c, sigma, x, pi):
    
    dim = c.shape[0]
    
    #responsibility = pi * (1 / ((np.sqrt(2 * np.pi) * sigma)**dim)) * np.exp((-1/(sigma**2)) * ((c - x)**2).sum()) 
    responsibility = pi * sp.stats.multivariate_normal(mean=c, cov=np.array([[sigma**2, 0],[0, sigma**2]])).pdf(x)
    
    return responsibility
```

```python
def assignment_step_softv2(centroids, xs, sigmas, pis, beta=1):
    
    assignments = []
    
    distance_to_centroids = np.zeros((xs.shape[0], centroids.shape[0]))
    
    for x_id, x in enumerate(xs):
        for c_id, c in enumerate(centroids):
            sigma = sigmas[c_id]
            pi = pis[c_id]
            distance_to_centroids[x_id, c_id] = beta * gaussian_dist(c, sigma, x, pi)
     
    normed_distances = distance_to_centroids / distance_to_centroids.sum(axis=1).reshape(-1, 1)
    normed_distances[np.isnan(normed_distances)] = 0
    
    return normed_distances
    
```

```python
def update_step_softv2(xs, assignments, n, curr_sigmas, curr_centroids):
    
    all_centroid = []
    all_sigma = []
    
    for i in range(n):
        centroid_resps = assignments[:, i]
        
        curr_sigma = curr_sigmas[i]
        curr_centroid = curr_centroids[i,:]
        
        total_resp = centroid_resps.sum()
        
        if total_resp == 0:
            all_centroid.append(curr_centroid)
            all_sigma.append(curr_sigma)
            print("no resp for cluster %s" % str(i))
        else:
            # m
            centroid = centroid_resps @ xs / total_resp

            # sigma
            sigma = (centroid_resps @ ((xs - centroid)**2).sum(axis=1)) / (xs.shape[1] * total_resp)
            
            if sigma == 0:
                sigma = 0.1
                
            all_centroid.append(centroid)
            all_sigma.append(np.sqrt(sigma))
    
    # pi
    pi = assignments.sum(axis=0) / assignments.sum()
    
    return np.row_stack(all_centroid), np.array(all_sigma), pi
```

```python
def run_simulation_softv2(xs, n_clusters, max_iter = 500, verbose=False):
    
    # We don't want to intialize two clusters to the same point
    unique_xs = np.unique(xs, axis=0)

    # initialize clusters
    centroids_idx = np.random.choice(np.arange(unique_xs.shape[0]), size=n_clusters, replace=False)
    
    new_centroids = np.random.uniform(0, 10, size=(n_clusters, 2))
    
    #new_centroids = unique_xs[centroids_idx]
    prev_centroids = np.zeros_like(new_centroids)

    sigmas = np.full(n_clusters, 0.5) #np.random.uniform(0, 1, size=n_clusters)
    pis = np.full(n_clusters, 1/n_clusters)
    
    assignments = assignment_step_softv2(new_centroids, xs, sigmas, pis)
    iteration = 0
    
    #print("assignment: \n", assignments, sep="")
    
    while (not np.isclose(new_centroids, prev_centroids, atol=1e-16, rtol=1e-14).all()) and (iteration < max_iter):
        if verbose: print(iteration, " >> new_centroids: ", new_centroids, "\n prev_centroids: ", prev_centroids, "\n sigmas: ", sigmas, "\n pis :", pis)
        prev_centroids = new_centroids
        new_centroids, sigmas, pis = update_step_softv2(xs, assignments, n_clusters, sigmas, prev_centroids)
        assignments = assignment_step_softv2(new_centroids, xs, sigmas, pis)
        if verbose: print("assignment:  \n", assignments, sep="")
        iteration += 1
    
    print(iteration)
    return assignments, new_centroids, np.column_stack([sigmas, sigmas])
```

```python
def gaussian_dist_v3(c, sigmas, x, pi):
    
    responsibility = pi * sp.stats.multivariate_normal(mean=c, cov=np.diag(sigmas**2)).pdf(x)

    return responsibility
```

```python
def assignment_step_softv3(centroids, xs, sigmas, pis):
    
    assignments = []
    
    distance_to_centroids = np.zeros((xs.shape[0], centroids.shape[0]))
    
    for x_id, x in enumerate(xs):
        for c_id, c in enumerate(centroids):
            sigma = sigmas[c_id]
            pi = pis[c_id]
            distance_to_centroids[x_id, c_id] = gaussian_dist_v3(c, sigma, x, pi)
    
    normed_distances = distance_to_centroids / distance_to_centroids.sum(axis=1).reshape(-1, 1)
    normed_distances[np.isnan(normed_distances)] = 1 / centroids.shape[0]
    
    return normed_distances
    
```

```python
def update_step_softv3(xs, assignments, n, curr_sigmas, curr_centroids):
    
    all_centroid = []
    all_sigma = []
    
    for i in range(n):
        centroid_resps = assignments[:, i]
        
        curr_sigma = curr_sigmas[i]
        curr_centroid = curr_centroids[i,:]
        
        total_resp = centroid_resps.sum()
        
        if total_resp == 0:
            all_centroid.append(curr_centroid)
            all_sigma.append(curr_sigma)
            print("no resp for cluster %s" % str(i))
        else:
            # m
            centroid = centroid_resps @ xs / total_resp

            # sigma
            sigma = (centroid_resps @ ((xs - centroid)**2)) / (total_resp)
            sigma[np.isclose(sigma, 0)] = 0.01
                
            all_centroid.append(centroid)
            all_sigma.append(np.sqrt(sigma))
    
    # pi
    pi = assignments.sum(axis=0) / assignments.sum()
    
    return np.row_stack(all_centroid), np.row_stack(all_sigma), pi
```

```python
def run_simulation_softv3(xs, n_clusters, max_iter = 500, verbose=False):
    
    # We don't want to intialize two clusters to the same point
    unique_xs = np.unique(xs, axis=0)

    # initialize clusters
    centroids_idx = np.random.choice(np.arange(unique_xs.shape[0]), size=n_clusters, replace=False)
    #new_centroids = unique_xs[centroids_idx]
    
    new_centroids = np.random.uniform(0, 10, size=(n_clusters, 2))
    prev_centroids = np.zeros_like(new_centroids)

    sigmas = np.full_like(new_centroids, 0.5) #np.random.uniform(0, 1, size=n_clusters)
    pis = np.full(n_clusters, 1/n_clusters)
    
    assignments = assignment_step_softv3(new_centroids, xs, sigmas, pis)
    iteration = 0
    
    
    while (not np.isclose(new_centroids, prev_centroids, atol=1e-16, rtol=1e-14).all()) and (iteration < max_iter):
        if verbose: print(iteration, " >> new_centroids: ", new_centroids, "\n prev_centroids: ", prev_centroids, "\n sigmas: ", sigmas, "\n pis :", pis)
        prev_centroids = new_centroids
        new_centroids, sigmas, pis = update_step_softv3(xs, assignments, n_clusters, sigmas, prev_centroids)
        assignments = assignment_step_softv3(new_centroids, xs, sigmas, pis)
        if verbose: print("assignment:  \n", assignments, sep="")
        iteration += 1
    
    print(iteration)
    return assignments, new_centroids, sigmas
```

# Soft K-means clustering v4 - non-Axis-aligned cluster

```python
def gaussian_dist_v4(c, cov, x, pi):
    
    try:
        responsibility = pi * sp.stats.multivariate_normal(mean=c, cov=cov).pdf(x)
    except np.linalg.LinAlgError as err:
        responsibility = pi * sp.stats.multivariate_normal(mean=c, cov=np.array([[0.01, 0],[0, 0.01]])).pdf(x)

    return responsibility
```

```python
def assignment_step_softv4(centroids, xs, covs, pis):
    
    assignments = []
    
    distance_to_centroids = np.zeros((xs.shape[0], centroids.shape[0]))
    
    for x_id, x in enumerate(xs):
        for c_id, c in enumerate(centroids):
            cov = covs[c_id]
            pi = pis[c_id]
            distance_to_centroids[x_id, c_id] = gaussian_dist_v4(c, cov, x, pi)
    
    normed_distances = distance_to_centroids / distance_to_centroids.sum(axis=1).reshape(-1, 1)
    normed_distances[np.isnan(normed_distances)] = 1 / centroids.shape[0]
    
    return normed_distances
```

```python
def update_step_softv4(xs, assignments, n, curr_sigmas, curr_centroids):
    
    all_centroid = []
    all_sigma = []
    
    for i in range(n):
        centroid_resps = assignments[:, i]
        
        curr_sigma = curr_sigmas[i]
        curr_centroid = curr_centroids[i,:]
        
        total_resp = centroid_resps.sum()
        
        if total_resp == 0:
            all_centroid.append(curr_centroid)
            all_sigma.append(curr_sigma)
            print("no resp for cluster %s" % str(i))
        else:
            # m
            centroid = centroid_resps @ xs / total_resp

            # diag terms
            w_dist = xs - centroid
            
            # cov matrix
            cov_matrix = np.cov(w_dist.T, aweights=centroid_resps)    
            cov_matrix = np.nan_to_num(cov_matrix)
                
            all_centroid.append(centroid)
            all_sigma.append(cov_matrix)
    
    # pi
    pi = assignments.sum(axis=0) / assignments.sum()
    
    return np.row_stack(all_centroid), np.stack(all_sigma), pi
```

```python
def run_simulation_softv4(xs, n_clusters, max_iter = 500, verbose=False):
    
    # We don't want to intialize two clusters to the same point
    unique_xs = np.unique(xs, axis=0)

    # initialize clusters
    centroids_idx = np.random.choice(np.arange(unique_xs.shape[0]), size=n_clusters, replace=False)
    #new_centroids = unique_xs[centroids_idx]
    
    new_centroids = np.random.uniform(0, 10, size=(n_clusters, 2))
    prev_centroids = np.zeros_like(new_centroids)

    sigmas = np.full_like(new_centroids, 0.5) #np.random.uniform(0, 1, size=n_clusters)
    pis = np.full(n_clusters, 1/n_clusters)
    
    assignments = assignment_step_softv4(new_centroids, xs, sigmas, pis)
    iteration = 0
    
    
    while (not np.isclose(new_centroids, prev_centroids, atol=1e-16, rtol=1e-14).all()) and (iteration < max_iter):
        if verbose: print(iteration, " >> new_centroids: ", new_centroids, "\n prev_centroids: ", prev_centroids, "\n sigmas: ", sigmas, "\n pis :", pis)
        prev_centroids = new_centroids
        new_centroids, sigmas, pis = update_step_softv4(xs, assignments, n_clusters, sigmas, prev_centroids)
        assignments = assignment_step_softv4(new_centroids, xs, sigmas, pis)
        if verbose: print("assignment:  \n", assignments, sep="")
        iteration += 1
    
    print(iteration)
    return assignments, new_centroids, sigmas
```
